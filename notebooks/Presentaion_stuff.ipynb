```python
import os
import multiprocessing
import gab
import numpy as np
import pandas
import json

num_processes = 112

filesDir = '../processed_data_sematic_eval/'

preDir = '../users_info/'
postDir = '../scripts/scaper/users_info/'
```

```python
targets = []
for e in os.scandir(filesDir):
    if e.name.startswith('post'):
        targets.append(e.path)
```

```python
def toDF(path):
    with open(path) as f:
        return pandas.DataFrame([json.loads(l) for l in f])
```

```python
with multiprocessing.Pool(16) as pool:
    dfs = pool.map(toDF, targets)
```

```python
df = pandas.concat(dfs)
```

```python
df['post'] = df['created_at'].apply(lambda x : True if x.startswith('2019') or x.startswith('2018-1') else False)
```

```python
df_p = df[df['post']]
```

```python
df_b = df[df['post'] != True]
```

```python
def loadTargets(dat):
    i, path = dat
    if i % 1000 == 0:
        print(f"{i} loading {path}".ljust(70).replace('\n', ''), end = '\r')
    try:
        A = gab.GabAccount(path)
    except gab.BadFileError:
        return None
    return A

def prepDirPath(path):
    loadPaths = []
    for e in os.scandir(path):
        if e.name.endswith('json.gz'):
            loadPaths.append(e.path)
    return loadPaths
```

```python
loadPaths = prepDirPath(postDir)
```

```python
with multiprocessing.Pool(num_processes) as pool:
    newTargets = pool.map(loadTargets, enumerate(loadPaths))
```
